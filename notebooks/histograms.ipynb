```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to compute histogram for a single image
def compute_histogram(image):
    # Compute histogram
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    return hist.flatten()  # Flatten the histogram

# Function to load images from a folder and compute histograms
def load_and_compute_histograms(folder_path):
    histograms = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".jpg"):
            image_path = os.path.join(folder_path, filename)
            # Load image
            image = cv2.imread(image_path)
            if image is not None:
                # Convert to grayscale if necessary
                # Compute histogram
                hist = compute_histogram(image)
                histograms.append(hist)
    return histograms
```

```python
# Paths to the directories containing Cat and Dog folders
base_path = "/Users/blanchepenafiel/Desktop/Dataset"
train_path= os.path.join(base_path, "Train")
cat_path = os.path.join(base_path, "Train", "Cat")
dog_path = os.path.join(base_path, "Train", "Dog")
```

```python
# Load and compute histograms for cat and dog images
cat_histograms = load_and_compute_histograms(cat_path)
dog_histograms = load_and_compute_histograms(dog_path)
```

```python
# Plot histograms (optional)
plt.figure()
plt.plot(np.mean(cat_histograms, axis=0), label='Cat')  # Plot the mean histogram for cats
plt.plot(np.mean(dog_histograms, axis=0), label='Dog')  # Plot the mean histogram for dogs
plt.title('Mean Histograms of Pixel Intensities')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Plot histograms separately for cats and dogs
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(np.mean(cat_histograms, axis=0), label='Cat', color='blue')
plt.title('Mean Histogram of Pixel Intensities for Cats')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')


plt.subplot(1, 2, 2)
plt.plot(np.mean(dog_histograms, axis=0), label='Dog', color='orange')
plt.title('Mean Histogram of Pixel Intensities for Dogs')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')


plt.show()
```

```python
# Plot histograms separately for cats and dogs
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(range(len(np.mean(cat_histograms, axis=0))), np.mean(cat_histograms, axis=0), color='blue', alpha=0.7, label='Cat')
plt.title('Mean Histogram of Pixel Intensities for Cats')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 2, 2)
plt.bar(range(len(np.mean(dog_histograms, axis=0))), np.mean(dog_histograms, axis=0), color='orange', alpha=0.7, label='Dog')
plt.title('Mean Histogram of Pixel Intensities for Dogs')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Plot histograms overlapped for cats and dogs
plt.figure(figsize=(8, 6))

# Plot histogram for cats
plt.bar(range(len(np.mean(cat_histograms, axis=0))), np.mean(cat_histograms, axis=0), color='blue', alpha=0.5, label='Cat')

# Plot histogram for dogs
plt.bar(range(len(np.mean(dog_histograms, axis=0))), np.mean(dog_histograms, axis=0), color='orange', alpha=0.5, label='Dog')

plt.title('Mean Histograms of Pixel Intensities')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

The mean histograms pixel intensities are almost the sme which suggests that there might not be a siggnificant difference in the pixel intensity which alone might not be enough to distinguish between cats and dogs.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate example data (replace with your actual data)
cat_data = np.random.normal(loc=0, scale=1, size=1000)
dog_data = np.random.normal(loc=1, scale=1, size=1000)

# Adjust bin sizes for the histograms
bin_edges = np.linspace(min(min(cat_data), min(dog_data)), max(max(cat_data), max(dog_data)), num=20)

# Plot normalized histograms
plt.figure(figsize=(10, 6))

plt.hist(cat_data, bins=bin_edges, density=True, alpha=0.5, label='Cat', color='blue')
plt.hist(dog_data, bins=bin_edges, density=True, alpha=0.5, label='Dog', color='orange')

plt.title('Normalized Histograms of Cat and Dog Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)

plt.show()
```
